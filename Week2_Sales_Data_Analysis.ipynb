```python
import pandas as pd
# 1. Load the dataset
df = pd.read_csv('sales_data.csv')

# 2. Quick Overview
print("--- Data Info ---")
print(df.info())

print("\n--- Summary Statistics ---")
print(df.describe())

# 3. Total Sales by Product Category
category_sales = df.groupby('Product_Category')['Sales_Amount'].sum().sort_values(ascending=False)
print("\n--- Sales by Category ---")
print(category_sales)

# 4. Total Sales by Region
region_sales = df.groupby('Region')['Sales_Amount'].sum().sort_values(ascending=False)
print("\n--- Sales by Region ---")
print(region_sales)

# 5. Simple Profit Calculation (Example)
# Assuming Profit = Sales_Amount - (Unit_Cost * Quantity_Sold)
df['Profit'] = df['Sales_Amount'] - (df['Unit_Cost'] * df['Quantity_Sold'])
total_profit = df['Profit'].sum()
print(f"\nTotal Calculated Profit: ${total_profit:,.2f}")
```
